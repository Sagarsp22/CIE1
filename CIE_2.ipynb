```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = sns.load_dataset("titanic")
df
```

```python
df.info()
```

```python
df.describe()
```

```python
print("Age - Mean:", df['age'].mean(), "Median:", df['age'].median(),"Mode:", df['age'].mode())
print("Fare - Mean:", df['fare'].mean(), "Median:",df['fare'].median(),"Mode:", df['fare'].mode())
```

```python
df['survived'] = df['survived'].replace({1: 'survived', 0: 'not survived'})
df['survived'].value_counts().plot(kind='bar')
plt.title('survived')
plt.show()
```

```python
df['pclass'].value_counts().sort_index().plot(kind='bar')
plt.title('PClass')
plt.show()
```

```python
print("missing values:")
df.isnull().sum()
df.fillna(df.mean(numeric_only=True))
df.isnull().sum()
```

```python
mean = np.mean(df['age'])
std = np.std(df['age'])
print(mean)
print(std)

threshold = 3
outlier = []
for i in df['age']:
  z = (i-mean)/std
  if z > threshold:
    outlier.append(i)
print(outlier)
sns.boxplot(x=df['age'])
plt.title('outliers')
plt.show()
```
