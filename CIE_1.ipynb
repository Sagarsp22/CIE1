```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('mtcars.csv')
plt.hist(df['mpg'],bins=10,edgecolor='b')
plt.title("Histogram of MPG")
plt.xlabel("MPG")
plt.ylabel("Frequency")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('mtcars.csv')
plt.scatter(df['mpg'],df['wt'])
plt.title("scatter of WT")
plt.xlabel("MPG")
plt.ylabel("Frequency")
plt.show()
```
